```python
import os
from PIL import Image
import pytesseract
import csv

# Directory where your image files are located
image_directory = '/content/'

# List of image file extensions you want to process
image_extensions = ['.jpg', '.jpeg', '.png']

# Initialize a list to store the extracted texts
extracted_texts = []

# Loop through the image files and perform OCR
for root, dirs, files in os.walk(image_directory):
    for image_file in files:
        if any(image_file.lower().endswith(ext) for ext in image_extensions):
            # Open the image
            image_path = os.path.join(root, image_file)
            image = Image.open(image_path)

            # Perform OCR on the image
            text = pytesseract.image_to_string(image)

            # Append the extracted text to the list
            extracted_texts.append({'Image File': image_file, 'Extracted Text': text})

# Determine the path to save the CSV file in the same folder as the images
csv_file_path = os.path.join(image_directory, 'extracted_texts.csv')

# Write the extracted texts to a CSV file
with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = ['Image File', 'Extracted Text']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for item in extracted_texts:
        writer.writerow(item)

# Print a message when the process is complete
print("Text extraction and CSV export completed.")
```

```python
import csv
from PIL import Image
import pytesseract
import os

# Create a list to store extracted text and image filenames
extracted_data = []

# Define a function to check if an image is valid
def is_valid_image(img):
    try:
        # Attempt to open the image to check its validity
        img.verify()
        return True
    except Exception:
        return False

# Loop through the images in your working directory and run OCR on each image
for image_file in os.listdir():
    if image_file.lower().endswith(('.jpg', '.jpeg', '.png')) and is_valid_image(Image.open(image_file)):
        try:
            image = Image.open(image_file)
            text = pytesseract.image_to_string(image)

            # Append the extracted text and image filename to the list
            extracted_data.append({'Image File': image_file, 'Extracted Text': text})
        except Exception as e:
            print(f"Error processing {image_file}: {str(e)}")

# Save the data to "ml.csv"
ml_csv_file_name = 'ml.csv'
with open(ml_csv_file_name, 'w', newline='') as ml_csvfile:
    fieldnames = ['Image File', 'Extracted Text']
    ml_writer = csv.DictWriter(ml_csvfile, fieldnames=fieldnames)
    ml_writer.writeheader()
    for item in extracted_data:
        ml_writer.writerow(item)
```
